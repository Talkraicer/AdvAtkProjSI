```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x

# Calculate MSE between y1 and y2
mse = np.mean((y1 - y2) ** 2)

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot data on each subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine Function')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine Function')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tangent Function')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Linear Function')

# Add MSE as an annotation
# Position the text between the first row of subplots
fig.text(0.25, 0.3, f'MSE between Sine and Cosine: {mse:.2f}', ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.5), rotation=90)

fig.text(0.75, 0.3, f'MSE between Sine and Cosine: {mse:.2f}', ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.5), rotation=90)

fig.text(0.5, 0.5, f'MSE between Sine and Cosine: {mse:.2f}', ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.5), rotation=300)

fig.text(0.5, 0.75, f'MSE between Sine and Cosine: {mse:.2f}', ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))


fig.text(0.5, 0.25, f'MSE between Sine and Cosine: {mse:.2f}', ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
